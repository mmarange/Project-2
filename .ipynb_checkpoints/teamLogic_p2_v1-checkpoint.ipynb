```python
#Reading csv files
income_raw_data = pd.read_csv("data/CAINC30_ALL_AREAS.csv", encoding = "ISO-8859-1")
income_raw_data
```

```python
#Analyse income_raw 
#Create dataframe from 1997 to 2019
#Drop irrelevant columns, 

years_dropped = list(range(1969,2000))
income_2000_2019 = income_raw_data

for year in years_dropped:
    income_2000_2019 = income_2000_2019.drop(columns = str(year))
income_2000_2019 = income_2000_2019.drop(columns = ['TableName', 'LineCode', 'GeoFIPS', 'Region', 'IndustryClassification'])
#income_df = income_1997_2019.drop(columns = ["GeoFips"])
income_2000_2019.head(100)
```

```python
income_2000_2019.shape
income_2000_2019.columns.tolist()
```

```python
income_2000_2019["County"] = ""
income_2000_2019["State"] = ""
income_2000_2019 = income_2000_2019[[
 'GeoName',
 'County',
 'State',
 'Description',
 'Unit',
 '2000',
 '2001',
 '2002',
 '2003',
 '2004',
 '2005',
 '2006',
 '2007',
 '2008',
 '2009',
 '2010',
 '2011',
 '2012',
 '2013',
 '2014',
 '2015',
 '2016',
 '2017',
 '2018',
 '2019'
 ]]

income_2000_2019.head(3)
```

```python
record_len = len(income_2000_2019)

for i in range(record_len-1):
    income_2000_2019.iloc[i,1] = income_2000_2019.iloc[i,0].split(",")[0] 
    try:
        income_2000_2019.iloc[i,2] = income_2000_2019.iloc[i,0].split(",")[1]  
    except: print(f'record {i} not a county')

example = income_2000_2019.iloc[99,1].split(",")
income_2000_2019.head()
#example
```

```python
nan_value = float("NaN")
income_2000_2019.replace("", nan_value, inplace=True)
income_2000_2019_df = income_2000_2019.dropna(axis=0, how='any', subset=['State'])

income_2000_2019["Description"]=income_2000_2019["Description"].str.strip()
income_2000_2019["Unit"]=income_2000_2019["Unit"].str.strip()



income_2000_2019_df.head()
```

```python
income_2000_2019_df_FLT = income_2000_2019_df[(income_2000_2019_df["Description"] == "Population (persons) 3/") |
                                             (income_2000_2019_df["Description"] == "Per capita personal income 4/") |
                                             (income_2000_2019_df["Description"] == "Per capita net earnings 4/")|
                                             (income_2000_2019_df["Description"] == "Total employment (number of jobs)")|
                                             (income_2000_2019_df["Description"] == "Average earnings per job (dollars)")|
                                             (income_2000_2019_df["Description"] == "Average wages and salaries")]
income_2000_2019_df_FLT.head(15)
#states_mean_df[states_mean_df["Description"] == 'Per capita personal income (dollars) 2/']
```

```python
import geopandas as gpd
import pandas as pd
import re
import json
```

```python
zillow_df = pd.read_csv('zillow_home_values.csv')
```

```python
zillow_df.head()
```

```python
zillow_gdf = gpd.GeoDataFrame(zillow_df.copy())
zillow_gdf.head()
```

```python
with open('geocounty.geojson') as geojsonfile:
    county_json = json.load(geojsonfile)
```

```python
county_gdf = gpd.read_file("us-county-boundaries.geojson")
county_gdf.head()
```

```python
print(f'GeoJSON {len(county_gdf)}')
print(f'Zillow {len(zillow_gdf)}')
```

```python
# merged_geojson = zillow_gdf.merge(county_gdf, left_on=['RegionName','State'], right_on=['namelsad','stusab'])
merged_geojson = county_gdf.merge(zillow_gdf, left_on=['RegionName','State'], right_on=['namelsad','stusab'], how='left')

# @VERA: Search for null columns in the zillow columns to identify non-matches
```

```python
print(f'Merged Zillow w/ GeoJSON {len(merged_geojson)}')
```
